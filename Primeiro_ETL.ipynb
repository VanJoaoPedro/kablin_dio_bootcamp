## Primeiro ETL

### Extract

Extraindo dados de clientes de um arquivo Json.

```python
import json
with open('clientes.json', 'r', encoding='utf-8') as arquivo:
  dados = json.load(arquivo)

lista_clientes = []

for item in dados:
  lista_clientes.append([
      item["name"],
      item["account"]["number"],
      item["account"]["agency"],
      item["account"]["balance"],
      item["account"]["limit"],
      item["card"]["number"],
      item["card"]["limit"]
  ])
for linha in lista_clientes:
  print(linha)
```

### Transform
Transformando os dados. Inserindo comentarios de incentivo ao investimento utilizando o gemini e convertendo para um dataframe no pandas.

```python
import os
from google import genai

# Configure sua API Key do Google AI Studio
os.environ["GEMINI_API_KEY"] = "AIzaSyAUmLxNs_A9eqxVcMPQ01J30hJn13Y3lNo"

# Inicializa o cliente do Gemini
client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])

def generate_ai_news(user):
    # O método generate_content é o equivalente ao Chat Completions
    # Usamos o parâmetro config para definir as instruções de sistema
    response = client.models.generate_content(
        model="gemini-3.1-flash-lite-preview",
        config={
            "system_instruction": "Você é um especialista em marketing bancário."
        },
        contents=f"Crie uma mensagem para {user[0]} sobre a importância dos investimentos (máximo de 100 caracteres)"
    )

    # No Gemini, acessamos o texto diretamente pela propriedade .text
    return response.text.strip().replace('"', '')

# Loop de processamento
for cliente in lista_clientes:
    news = generate_ai_news(cliente)
    print(news)
    cliente.append(news)
print(lista_clientes)
```
